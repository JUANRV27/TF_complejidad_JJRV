```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
#Prueba correcta
# Cargar los archivos CSV
nodes_df = pd.read_csv('lima_streets_nodes.csv')
edges_df = pd.read_csv('lima_streets_edges.csv')

# Crear el grafo
G = nx.Graph()

# Añadir nodos al grafo
for _, row in nodes_df.iterrows():
    G.add_node(row['node_id'], pos=(row['x'], row['y']))

# Añadir aristas al grafo con la distancia como peso
for _, row in edges_df.iterrows():
    G.add_edge(row['node1'], row['node2'], weight=row['distance'])

# Extraer posiciones de los nodos
pos = nx.get_node_attributes(G, 'pos')
x_nodes = [pos[n][0] for n in G.nodes()]  # Coordenadas X
y_nodes = [pos[n][1] for n in G.nodes()]  # Coordenadas Y
node_ids = list(G.nodes())  # IDs de nodos

# Dibujar el grafo
plt.figure(figsize=(10, 10))
nx.draw(G, pos, node_size=10, node_color='blue', edge_color='gray', with_labels=False)
plt.title('Grafo de Calles de Lima')

# Crear aristas para el gráfico
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x += [x0, x1, None]
    edge_y += [y0, y1, None]
    
edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.5, color='#888'),
    hoverinfo='none',
    mode='lines')

node_trace = go.Scatter(
    x=x_nodes, y=y_nodes,
    mode='markers',
    marker=dict(
        size=10,
        color='blue',
        line=dict(width=1)),
    text=node_ids,  # Texto que aparece al pasar el cursor
    hoverinfo='text'
)

fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title='Grafo de Calles de Lima',
                    showlegend=False,
                    hovermode='closest',
                    margin=dict(b=0, l=0, r=0, t=40),
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                )

fig.show()
plt.show()

# Función para encontrar la ruta más corta usando Dijkstra
def shortest_path_dijkstra(graph, start, end):
    return nx.dijkstra_path(graph, start, end, weight='weight')

# Definir nodos de ejemplo para encontrar la ruta más corta
start_node = nodes_df.iloc[0]['node_id']
end_node = nodes_df.iloc[10]['node_id']

# Encontrar la ruta más corta entre los dos nodos
shortest_path = shortest_path_dijkstra(G, start_node, end_node)

input(start_node)
input(end_node) 
print("Ruta más corta:", shortest_path)
```
